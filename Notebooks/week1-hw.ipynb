```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
style.use('bmh')
#Not to show warning messages (to keep the notebook clean)
import warnings
warnings.filterwarnings('ignore')
```

**3**

```python
#time periods
size=2000
#Arrival
mean=3
arrival = np.random.poisson(mean, size)
#Service
mean=4
service = np.random.poisson(mean, size)
#
queue = np.zeros(size+1)

counter = 0

for i in range(size):
    if ((arrival[i] - service[i]) + queue[i]) > 0:
        queue[i+1] = max(0,(arrival[i] - service[i]) + queue[i])
    if queue[i] == 0:
        counter = counter + 1


plt.figure(figsize=(14,5))
plt.hist( queue )
plt.xlabel('$n$')
plt.ylabel('Size')
plt.title('Queue')
plt.grid(True) #grid
plt.show() 

print('Idle periods:', counter)
print('Idle periods ratio:', counter/size)
```
